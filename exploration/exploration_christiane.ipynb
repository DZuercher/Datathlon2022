Looks like generator_speed is very similar to rotor_speed, this might mean we can drop one of these as features. 
ERROR_5 seems to be connected to high power, and high wind_speed. If there is no error, blade_angle_avg tends to be low, if it is high, this can indicate an error. Similarly, if cosphi is low, this can also indicate an error. 
In total, promising features seem to be blade_angle_avg, hydraulic_pressure, cosphi, wind_speed, rotor_speed, power. For wind_direction and nacelle_direction it is less clear.
